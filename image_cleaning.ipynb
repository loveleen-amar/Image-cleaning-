```
"""here we add different noise to every column(784) of row(60000) and than predict to do image clean or not"""
```

```
from tensorflow.keras.datasets import mnist
```

```
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```
x_train=x_train.reshape(-1,784)
```

```
import numpy as np
```

```
noise1=np.random.randint(0,100,(len(x_train),784))
```

```
noise2=np.random.randint(0,100,(len(x_test),784))
```

```
x_train_mod=x_train+noise1
```

```
x_test=x_test.reshape(-1,784)
```

```
x_test_mod=x_test+noise2
```

```
y_train_mod=x_train
```

```
y_test_mod=x_test
```

```
from sklearn.neighbors import KNeighborsClassifier
```

```
knn_clf=KNeighborsClassifier(n_neighbors=10)
```

```
knn_clf.fit(x_train_mod,y_train_mod)
```

```
clean_digit=knn_clf.predict([x_test_mod[0]])
```

```
clean_digit.shape
```

```
import matplotlib as mpl
import matplotlib.pyplot as plt

some_digit=clean_digit
some_digit_image=some_digit.reshape(28,28)

plt.imshow(some_digit_image,cmap=mpl.cm.binary,interpolation="nearest")
plt.axis("off")
plt.show()
```

```
"""the above image is the clean image of the prediction after don cleaning of original image"""
```
